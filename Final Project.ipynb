```python
#Data Preprocessing
#Encoding'smoker'column
data.replace({'smoker':{'yes':0,'no':1}}, inplace=True)
```

```python
#Encoding'sex'column
data.replace({'sex':{'male':0,'female':1}}, inplace=True)
```

```python
#Encoding'region'column
data.replace({'region':{'southeast':1,'southwest':2,'northeast':3,'northwest':4}}, inplace=True)
```

```python
data.head
```

```python
# Spliting The Data into Features and Target
X = data.drop('charges', axis = 1)
y = data['charges']
```

```python
corr = data.corr()
plt.figure(figsize= (10, 10))
sns.heatmap(corr, annot=True, mask=np.triu(corr))
plt.title("Correlation of Variables")
plt.show()
```

Modeling

```python
# Spliting into Train data and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression   ## importing the model library

lr=LinearRegression()  ## importing the model object
lr.fit(X_train,y_train) ## fitting the training data

X_test_pred=lr.predict(X_test)
```

```python
print(X_test_pred)
```

```python
X_train_pred=lr.predict(X_train)
print(X_train_pred)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
training_num=r2_score(y_train,X_train_pred)
print('The Linear Regression training is :',training_num)
```

```python
testing_num=r2_score(y_test,X_test_pred)
print('The Linear Regression testing is :',testing_num)
```

```python
print('MSE of training is :',mean_squared_error(y_train,X_train_pred))
print('MSE of testing is:',mean_squared_error(y_test,X_test_pred))
```

```python
print('R^2 of training is :', r2_score(y_train,X_train_pred))
print('R^2 of testing is:', r2_score(y_test,X_test_pred))
```

```python
print('RMSE of training is :', np.sqrt(mean_squared_error(y_train,X_train_pred)))
print('RMSE of testing is:', np.sqrt(mean_squared_error(y_test,X_test_pred)))
```
